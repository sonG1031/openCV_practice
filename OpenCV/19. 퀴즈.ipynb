## OpenCV를 이용하여 가로로 촬영된 영상을 세로로 회전하는 프로그램을 작성하시오

### 조건
1. 회전 : 시계 반대방향으로 90도
1. 재생속도 (FPS) : 원본 x 4배
1. 출력 파일명: city_output.avi(코덱 : DIVX)

```python
import cv2
cap = cv2.VideoCapture('city.mp4')

fourcc = cv2.VideoWriter_fourcc(*'DIVX')
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS) * 4

out = cv2.VideoWriter('city_output.avi', fourcc, fps, (height, width)) # 가로와 세로가 바뀌는 것 주의!

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    rotate_270 = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE) 
    out.write(rotate_270)
    cv2.imshow('video', rotate_270)
    if cv2.waitKey(1) == ord('q'):
        break

out.release()
cap.release()
cv2.destroyAllWindows()
```
